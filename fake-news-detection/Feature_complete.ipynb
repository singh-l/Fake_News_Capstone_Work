```python
import numpy as np
import pandas as pd
import itertools
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#Read the data
df=pd.read_csv('train.csv')
#Get shape and head
df.shape
```

```python
df['content']=df['title']+df['text']
df=df.dropna()
df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 
df['t_sentiment'] = (df['content']).map(lambda x:SentimentIntensityAnalyzer().polarity_scores(x)['compound'])
```

```python
df
```

```python
import string
count = lambda l1,l2: sum([1 for x in l1 if x in l2])

df['count_punct'] = (df.content).apply(lambda s: count(s, string.punctuation))
```

```python
df
```

```python
import textstat as ts
df['Readability'] = (df['content']).map(lambda x:ts.automated_readability_index(x))
```

```python
df
```

```python
df.to_csv('featured_data.csv')
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.loc[:, 'count_punct':'Readability'], df['label'], test_size=0.2, random_state=7)
```

```python
x_train
```

```python
# KNN, all features
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
y_predict = classifier.predict(x_test)
print(classification_report(y_test, y_predict))

#KNN, only count_punc
classifier.fit(x_train.loc[:,'count_punct':'count_punct'], y_train)
y_predict = classifier.predict(x_test.loc[:,'count_punct':'count_punct'])
print(classification_report(y_test, y_predict))

#KNN, only Readability
classifier.fit(x_train.loc[:,'Readability':'Readability'], y_train)
y_predict = classifier.predict(x_test.loc[:,'Readability':'Readability'])
print(classification_report(y_test, y_predict))

#KNN, only t_sentiemnt
classifier.fit(x_train.loc[:,'t_sentiment':'t_sentiment'], y_train)
y_predict = classifier.predict(x_test.loc[:,'t_sentiment':'t_sentiment'])
print(classification_report(y_test, y_predict))
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix


# SVM, all features

classifier = SVC()
classifier.fit(x_train, y_train)
y_predict = classifier.predict(x_test)
print(classification_report(y_test, y_predict))

#SVM, only count_punc
classifier.fit(x_train.loc[:,'count_punct':'count_punct'], y_train)
y_predict = classifier.predict(x_test.loc[:,'count_punct':'count_punct'])
print(classification_report(y_test, y_predict))

#SVM, only Readability
classifier.fit(x_train.loc[:,'Readability':'Readability'], y_train)
y_predict = classifier.predict(x_test.loc[:,'Readability':'Readability'])
print(classification_report(y_test, y_predict))

#SVM, only t_sentiment
classifier.fit(x_train.loc[:,'t_sentiment':'t_sentiment'], y_train)
y_predict = classifier.predict(x_test.loc[:,'t_sentiment':'t_sentiment'])
print(classification_report(y_test, y_predict))
```
